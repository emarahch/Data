# Questions after looking at the data

*   Is there a correlation between total number of abscences vs total average grade?
*   Is there a correlation between total number of self study hours vs total average grade?
* What is the correlation between career aspirations and total average grade?


```python
# Calculating the total average for each student


df['Total_Average'] = (df['math_score']+df['history_score']+df['physics_score']+df['chemistry_score']+df['biology_score']+df['english_score']+df['geography_score']) /7
df
```

```python
df.describe()
```

```python
careers=df["career_aspiration"]
plt.figure(figsize=(10, 6))
sns.countplot(data=careers, order=careers.value_counts().index)
plt.title('Distribution of Careers')
plt.xlabel('Frequency')
plt.ylabel('Careers')

```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x="absence_days", y="Total_Average")
plt.title('Absence vs Total average')
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x="weekly_self_study_hours", y="Total_Average")
plt.title('Weekly self study hours vs Total average')
```

```python
sns.boxplot(x=df['weekly_self_study_hours'])
plt.title('Distribution of Weekly Self-Study Hours')
plt.xlabel('Weekly Self-Study Hours')
```

```python
plt.figure(figsize=(10, 10))
sns.barplot(data=df, y="Total_Average", hue="part_time_job", errorbar=None);
plt.title('Part time job vs Total average')
```

```python
plt.figure(figsize=(10, 10))
sns.barplot(data=df, y="Total_Average", hue="extracurricular_activities",errorbar=None);
plt.title('Extracurricular activities vs Total average')
```

```python
plt.figure(figsize=(30, 10))
sns.barplot(data=df, x="career_aspiration", y="Total_Average", hue="career_aspiration",errorbar=None);
plt.title('Career aspirtations vs Total average')
```
